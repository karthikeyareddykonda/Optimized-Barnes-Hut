```python
AU = (149.6e6 )     # 149.6 million km
def random_gen(filename, num_points, massRange, xrange, yrange, zrange):
    with open(filename,'w') as file:
        file.write(str(num_points)+"\n")
        x_vals = np.linspace(xrange[0],xrange[1],num_points)
        y_vals = np.linspace(yrange[0],yrange[1],num_points)
        z_vals = np.linspace(zrange[0],zrange[1],num_points)
        masses = np.random.uniform(massRange[0],massRange[1],num_points)
        np.random.shuffle(x_vals)
        np.random.shuffle(y_vals)
        np.random.shuffle(z_vals)
        for i in range(num_points):
            x = x_vals[i]
            y = y_vals[i]
            z = z_vals[i]
            vx = np.random.uniform(low =-100,high = 100)
            vy = np.random.uniform(low =-100,high = 100) # in Km/s
            vz = np.random.uniform(low =-100,high = 100) # in Km/s
         
            m = masses[i]
            params = [x,y,z,vx,vy,vz,m]
       
            file.write(",".join(f"{v:.6e}" for v in params)+ "\n") # Does precision matter here ?
    



```

```python
num_points = np.arange(100000,1000000,100000)
mass_earth = 5.97219E+24
xrange = [-10*AU,10*AU]
yrange = [-10*AU,10*AU]
zrange = [-10*AU,10*AU]
massRange = [1* mass_earth, 4*mass_earth]
theta_vals = np.array([0.25,0.5,1.0,2.0,5.0,10.0])
timeStep = 86400 # in seconds
steps = 10 #  days # Warning larger timesteps is leading to memory limit breaches in baseline Implementations. Suspected positions becoming same during simulations
```
